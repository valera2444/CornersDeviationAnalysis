```python
fig, ax = plt.subplots(1,2, figsize=(20,10))
img = mpimg.imread(hist_means)
ax[0].imshow(img)
img1 = mpimg.imread(box_plot_corners_mean)
ax[1].imshow(img1)
```

These diagrams show distribution of mean deviations. We can see that there are a lot of large mean deviations, which means that there are a lot of 'extremal' errors in our model.

Also we can see, that model gives more 'extremely' wrong answers for rooms with 6 corners (wrt this is less examples with 6 corners than with 4) 
Also we can see, that deviations for 6 corners are bigger deviations for 4 corners, which also can be prooved by taking mean deviations: 11.74362500644729 for 4 corners and 19.771737260469134 for 6 corners.
